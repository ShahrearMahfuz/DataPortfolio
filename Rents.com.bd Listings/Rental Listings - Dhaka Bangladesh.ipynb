```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Datasets/raw_data-all_rental_listings(rent.com.bd).csv')
df.head(50)
```

```python
#Check for duplicated rows
df.duplicated().any()
```

```python
# check for null rows
df.isnull().any()
```

```python
# Extract property size from title for the rows
# that have empty values on the property size column

df.loc[df['property size'].isnull(),'property size']=df[df['property size'].isnull()]["title"].str.extract('(\d{3,4})')[0]
```

```python
#Standardize the 'property type' column 
df.loc[df['property type'].str.contains('Residential'),'property type']='Residential Apartment'
df.loc[df['property type'].str.contains('Commercial'),'property type']='Commercial Space'
df.loc[df['property type'].str.contains('Furnished'),'property type']='Residential Apartment'
df.head(50)

```

```python
#Remove unwanted characters and substrings from 'property size' column
df['property size']=df['property size'].str.replace('.','')
df['property size']=df['property size'].str.split('Sq Ft').str[0]
df=df.rename(columns={'property size':'property size(Sq Ft)'})
df.head(50)
```

```python
# Standardize parking spots column  
df['parking']=df['parking'].str.strip('ÂCcar Pparking')
df=df.rename(columns={'parking':'parking spots'})
#df.head(50)
```

```python
# Remove null values and unwanted strings in lift column
df['lift']=df['lift'].str.replace('available','')
df['lift']=df['lift'].str.replace('Available','')
```

```python
df.loc[df['lift'].isnull(),'lift']=df[df['lift'].isnull()].fillna(0)
df.head(50)
```

```python
# Clean the floor column and rename it
df['floor']=df['floor'].str.replace('Individual','Singular')
df['floor']=df['floor'].str.replace('Ground','G')
df['floor']=df['floor'].str.replace('–','&')
df['floor']=df['floor'].str.replace('available','')
df['floor']=df['floor'].str.replace('Available','')
df['floor']=df['floor'].str.replace('floor','')
df['floor']=df['floor'].str.replace('Floor','')
df['floor']=df['floor'].str.replace('st','')
df['floor']=df['floor'].str.replace('nd','')
df['floor']=df['floor'].str.replace('rd','')
df['floor']=df['floor'].str.replace('th','')
df=df.rename(columns={'floor':'floor available'})
```

```python
df.head(50)
```

```python
# Clean price column and remove unwanted substrings
df['price']=df['price'].str.split('BDT').str[1]
df['price']=df['price'].str.split('/').str[0]
df['price']=df['price'].str.replace(',','')
df['price'] = df['price'].str.split(n=2).str[:2].str.join(' ')
```

```python
# Remove null rows in price column
df = df[df['price'].notna()]
```

```python
# convert values with 'Cr' into full valued strings
priceCR=df.loc[df['price'].str.contains(' Cr'),'price']
priceCR=priceCR.str.replace('Cr.','Cr', regex=True)
priceCR=priceCR.str.replace('cr.','Cr', regex=True)
priceCR=priceCR.apply(lambda x:str( float(x.replace(' Cr', '')) * 10000000))
df.loc[df['price'].str.contains(' Cr'),'price']=priceCR
```

```python
# convert values with 'Lac' into full valued strings
priceLac=df.loc[df['price'].str.contains(' Lac'),'price']
priceLac=priceLac.apply(lambda x:str( float(x.replace(' Lac', '')) * 100000))
df.loc[df['price'].str.contains(' Lac'),'price']=priceLac
```

```python
# Convert the price column string values into integer values
df['price']=df['price'].str.extract('(\d+)')
df['price']=df['price'].astype(int)
df=df.rename(columns={'price':'price(BDT)'})
```

```python
df.head(50)
```

```python
# Remove unwanted substrings and standardize service_charge column
df['service_charge']=df['service_charge'].str.replace('N/A','')
df['service_charge']=df['service_charge'].str.replace('BDT','')
df['service_charge']=df['service_charge'].str.split('/').str[0]
df['service_charge']=df['service_charge'].str.replace(',','')
df['service_charge']=df['service_charge'].str.replace('.','')
df['service_charge']=df['service_charge'].replace([' Not fixed yet',' Not Fixed Yet',' Not Fixed yet'],'Not fixed yet')
df['service_charge']=df['service_charge'].replace(' At Actual','At Actual')

changeWords=['Including',' Including',' Including Rent',' Including with monthly rent',' Included with monthly rent',' Including With Rent',' Including with rent']
df.loc[df['service_charge'].isin(changeWords),'service_charge']="Including with rent"

df['service_charge']=df['service_charge'].fillna(0)


df=df.rename(columns={'service_charge':'service charge(BDT)'})
```

```python
df.head(40)
```

```python
# reduce multiple aliases of the same type of record to one record for preferred tennant column
df['preferred tennant'].unique()
```

```python
df['preferred tennant']=df['preferred tennant'].replace([' Corporate Office or MNC Offices',' Corporate or MNC Office','Corporate Office or MNC Offices.',' Corporate Office or MNC Offices.',' Corporate Office or MNC',' Corporate Office, MNC',' Corporate Office or MNC/ Restaurant'],'Corporate Office or MNC Office')
df['preferred tennant']=df['preferred tennant'].replace([' Corporate Office or Small Offices',' Corporate Office or Small Office'],'Corporate Office or Small Office')
df['preferred tennant']=df['preferred tennant'].replace(' Restaurant Brand',' Restaurant')
df['preferred tennant'].unique()
```

```python
# reduce multiple aliases of the same type of record to one record for interior column
sorted(df['interior'].unique())
```

```python
df['interior']=df['interior'].replace([' Full Furnished',
 ' Full-Furnished',
 ' Fully Furnished',
 ' Fully-Furnished',
 ' Furnished','Full-Furnished',
 'Fully-Furnished','Furnished'],'Full-Furnished')

df['interior']=df['interior'].replace([' Semi-Furnished',
 ' Semi-Unfurnished',
 ' Semi-furnished','Semi ',
 'Semi-Furnished'],'Semi-Furnished')

df['interior']=df['interior'].replace([' Non-Furnished',' Un-Furnished',
 ' UnFurnished',
 ' Unfurnished','Un',
 'Un ',
 'Un- Furnished',
 'Un-Furnished',
 'Un-furnished'],'Un-Furnished')

sorted(df['interior'].unique())
```

```python
# Change year built column to int data
df['year built']=df['year built'].astype(int)
```

```python
#Remove unwanted characters from garage size, front road size, common area columns
df['garage size']=df['garage size'].str.replace('.','')
df['garage size']=df['garage size'].str.replace('Sq Ft','')
df['garage size']=df['garage size'].str.replace(' Sq Ft','')
df['garage size']=df['garage size'].str.replace(' sq ft','')
df=df.rename(columns={'garage size':'garage size(Sq Ft)'})

df['front road size']=df['front road size'].str.replace('.','')
df['front road size']=df['front road size'].str.replace(' Ft','')
df['front road size']=df['front road size'].str.replace('Ft','')
df=df.rename(columns={'front road size':'front road size(Ft)'})

df['common area']=df['common area'].str.replace('.','')
df['common area']=df['common area'].str.replace('.','')
df['common area']=df['common area'].str.replace(' Sq Ft','')

df=df.rename(columns={'common area':'common area(Sq Ft)'})
```

```python
#Remove unwanted characters from bedrooms column

df['bedrooms']=df['bedrooms'].str.extract('(\d+)').astype(float)
df['bedrooms']=df['bedrooms'].fillna(0)
```

```python
#Remove unwanted characters from bathrooms column

df['bathrooms']=df['bathrooms'].str.extract('(\d+)').astype(float)
df['bathrooms']=df['bathrooms'].fillna(0)
```

```python
df=df.reset_index(drop=True)
```

```python
# Create new column postal code
df['postal code']=df['location'].str.extract('(\d{4})')
```
