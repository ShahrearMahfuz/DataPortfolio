# 22.Find unique values from Genre.

```python
lst=[]
for i in data['Genre']:
    lst.append(i.split(','))

lst #Shows genre in a 2d list
```

```python
list_1D=[]
for i in lst:
    for j in i:
        list_1D.append(j)
list_1D #Shows genre in 1d list
```

```python
uniq_genre=[]

for i in list_1D:
    if i not in uniq_genre:
        uniq_genre.append(i)

uniq_genre #Shows unique genre 
```

# 23.How many films of each genre were made?

```python
genre_dict={}
for i in uniq_genre:
        genre_dict[i]=list_1D.count(i)
```
